```python
#compute the degree centrality of dolphins
deg_cen = nx.degree_centrality(dolphins)
```

```python
#create a scatter plot of betweenness centrality and degree centrality
plt.scatter(list(bet_cen.values()), list(deg_cen.values()))
```

  ### Finding the nodes with the highest degree centrality

```python
def find_nodes_with_highest_deg_cent(dolphins):
    deg_cent= nx.degree_centrality(dolphins)
    max_dc = max(list(deg_cent.values()))
    nodes=set()
    for k, v in deg_cent.items():
        if v == max_dc:
            nodes.add(k)
    return nodes
```

```python
find_nodes_with_highest_deg_cent(dolphins)
```
